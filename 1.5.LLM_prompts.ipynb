# MinMaxScaler Mapping for All Questions

Create scalers for each unique scale range and map each question code to its appropriate scaler.

````python
from sklearn.preprocessing import MinMaxScaler
import pickle

# Define all unique scale ranges found in the survey
scale_ranges = {
    'scale_0_2': (0, 2),    # Q122-Q129 (Immigration effects: 0=Disagree, 1=Hard to say, 2=Agree)
    'scale_1_2': (1, 2),    # Q57, Q149, Q150, Q263 (Binary choices)
    'scale_1_3': (1, 3),    # Q42, Q43-Q45, Q56, Q111, Q173, Q200, Q209-Q212, Q213-Q216 (3-point scales)
    'scale_1_4': (1, 4),    # Q1-Q6, Q27-Q32, Q46, Q51-Q55, Q58-Q63, Q64-Q89, Q131, Q142-Q143, Q152-Q157, Q169-Q170, Q199, Q235-Q239, Q253, Q255-Q259 (4-point scales)
    'scale_1_5': (1, 5),    # Q33-Q41, Q47, Q121, Q201-Q208 (5-point scales)
    'scale_1_7': (1, 7),    # Q171 (Religious service attendance)
    'scale_1_8': (1, 8),    # Q172 (Prayer frequency)
    'scale_1_10': (1, 10),  # Q48-Q50, Q106-Q110, Q112, Q176, Q177-Q195, Q241-Q251 (10-point scales)
}

# Create a virtual dataset with all possible values for each scale
# This ensures the scaler knows the full range of each scale
virtual_data = {}
for scale_name, (min_val, max_val) in scale_ranges.items():
    # Create array with all possible integer values in the range
    virtual_data[scale_name] = np.arange(min_val, max_val + 1).reshape(-1, 1)

# Create and fit a MinMaxScaler for each scale range
scalers = {}
for scale_name, data in virtual_data.items():
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaler.fit(data)
    scalers[scale_name] = scaler
    print(f"{scale_name}: min={scaler.data_min_[0]}, max={scaler.data_max_[0]}, range={scaler.data_range_[0]}")

print(f"\nCreated {len(scalers)} scalers for different scale ranges")
````

````python
# Example: How to use the scalers

# Function to normalize a single question response
def normalize_response(question_code, response_value):
    """
    Normalize a response value for a given question code.
    
    Parameters:
    - question_code: str, e.g., 'Q1', 'Q57', etc.
    - response_value: int or float, the raw response value
    
    Returns:
    - float: normalized value between 0 and 1
    """
    scaler_name = question_to_scaler[question_code]
    scaler = scalers[scaler_name]
    normalized = scaler.transform([[response_value]])[0][0]
    return normalized

# Function to denormalize a value back to original scale
def denormalize_response(question_code, normalized_value):
    """
    Denormalize a value back to the original scale for a given question code.
    
    Parameters:
    - question_code: str, e.g., 'Q1', 'Q57', etc.
    - normalized_value: float between 0 and 1
    
    Returns:
    - float: value in original scale
    """
    scaler_name = question_to_scaler[question_code]
    scaler = scalers[scaler_name]
    denormalized = scaler.inverse_transform([[normalized_value]])[0][0]
    return denormalized

# Test with examples
print("Example normalizations:")
print("-" * 50)

examples = [
    ('Q1', 1, 'Family importance: Very important'),
    ('Q1', 4, 'Family importance: Not at all important'),
    ('Q57', 1, 'Trust: Most people can be trusted'),
    ('Q57', 2, 'Trust: Need to be careful'),
    ('Q177', 1, 'Never justifiable'),
    ('Q177', 10, 'Always justifiable'),
    ('Q122', 0, 'Immigration: Disagree'),
    ('Q122', 2, 'Immigration: Agree'),
]

for q_code, value, description in examples:
    normalized = normalize_response(q_code, value)
    print(f"{q_code} = {value} ({description})")
    print(f"  → Normalized: {normalized:.3f}")
    print(f"  → Back to original: {denormalize_response(q_code, normalized):.1f}")
    print()
````

````python
# Batch processing example: normalize all responses from a survey
def normalize_survey_responses(responses_dict):
    """
    Normalize all responses in a dictionary.
    
    Parameters:
    - responses_dict: dict, e.g., {'Q1': 2, 'Q2': 3, 'Q57': 1, ...}
    
    Returns:
    - dict: normalized responses with same keys
    """
    normalized = {}
    for q_code, value in responses_dict.items():
        if q_code in question_to_scaler and value is not None:
            normalized[q_code] = normalize_response(q_code, value)
        else:
            normalized[q_code] = None  # Keep None for missing values
    return normalized

# Example with sample responses
sample_responses = {
    'Q1': 1,    # Family: Very important
    'Q2': 2,    # Friends: Rather important
    'Q46': 2,   # Happiness: Rather happy
    'Q57': 1,   # Trust: Most people can be trusted
    'Q177': 3,  # Justifiability: Low (3 out of 10)
    'Q122': 2,  # Immigration: Agree
}

print("Sample Survey Responses:")
print("=" * 60)
normalized = normalize_survey_responses(sample_responses)

for q_code in sorted(sample_responses.keys()):
    original = sample_responses[q_code]
    norm = normalized[q_code]
    scale = question_to_scaler[q_code]
    print(f"{q_code}: {original} → {norm:.3f} (using {scale})")

print(f"\n✓ Normalized {len(normalized)} responses")
````
